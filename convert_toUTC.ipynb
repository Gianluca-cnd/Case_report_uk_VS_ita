```python
import pandas as pd
import numpy as np
```

```python
fpath = 'C:\\Users\\gg00642\\OneDrive - University of Surrey\\Desktop\\Actigraphy Sara'
```

```python
df = pd.read_csv(fpath + '\\1.0_sleep_calcs_all_joined.csv')
```

```python
# Define date ranges
date_ranges_location_0_subtract_2 = [
    ('2022-03-26', '2022-10-29'),
    ('2023-03-26', '2023-10-28'),
    ('2024-03-31', '2024-10-29')
]

date_ranges_location_1_subtract_1 = [
    ('2022-03-26', '2022-10-29'),
    ('2023-03-26', '2023-10-28'),
    ('2024-03-31', '2024-10-29')
]

date_ranges_location_0_subtract_1 = [
    ('2022-10-30', '2023-03-25'),
    ('2023-10-29', '2024-03-30'),
    ('2024-10-30', '2025-03-30')
]
```

```python
def adjust_value(row):
    date = row['date']
    location = row['location']
    
    if location == 0:
        for start, end in date_ranges_location_0_subtract_2:
            if pd.to_datetime(start) <= date <= pd.to_datetime(end):
                return row['sleep_start_decimal'] - 2, row['sleep_end_decimal'] - 2, row['midpoint_h'] - 2
        for start, end in date_ranges_location_0_subtract_1:
            if pd.to_datetime(start) <= date <= pd.to_datetime(end):
                return row['sleep_start_decimal'] - 1, row['sleep_end_decimal'] - 1, row['midpoint_h'] - 1
    elif location == 1:
        for start, end in date_ranges_location_1_subtract_1:
            if pd.to_datetime(start) <= date <= pd.to_datetime(end):
                return row['sleep_start_decimal'] - 1, row['sleep_end_decimal'] - 1, row['midpoint_h'] - 1
    
    return row['sleep_start_decimal'], row['sleep_end_decimal'], row['midpoint_h']

df['date'] = pd.to_datetime(df['date'])  # Convert 'date' column to datetime

df[['sleep_start_decimal_UTC', 'sleep_end_decimal_UTC', 'midpoint_h_UTC']] = df.apply(adjust_value, axis=1, result_type='expand')

print(df)
```

```python
df.to_csv(fpath + '\\1.1_sleep_calcs_all_joined_UTC.csv', index=False)
```
