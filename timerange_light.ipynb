```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
fpath = 'C:\\Users\\gg00642\\OneDrive - University of Surrey\\Desktop\\Actigraphy Sara'
```

```python
df = pd.read_csv(fpath + '\\light_2022.csv')
```

```python
df2 = pd.read_csv(fpath + '\\light_2023.csv')
```

```python
df3 = pd.read_csv(fpath + '\\light_2024.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])
```

```python
df['hour'] = df['date'].dt.hour
df2['hour'] = df2['date'].dt.hour
df3['hour'] = df3['date'].dt.hour
```

```python
# overlay all years
plt.figure(figsize=(10, 6))
plt.scatter(df['hour'], df['whitelight'], alpha=0.3, label='2022')
plt.scatter(df2['hour'], df2['whitelight'], alpha=0.3, label='2023')
plt.scatter(df3['hour'], df3['whitelight'], alpha=0.3, label='2024')
plt.xlabel('Hour')
plt.ylabel('Whitelight')
plt.title('Whitelight vs. Hour')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(2000))
plt.xlim(-0.5, 24)
plt.ylim(-800, 20000)
plt.legend()
plt.show()
```
