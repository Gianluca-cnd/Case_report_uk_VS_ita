```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
fpath = 'C:\\Users\\gg00642\\OneDrive - University of Surrey\\Desktop\\Actigraphy Sara'
```

```python
df = pd.read_csv(fpath + '\\b_2022.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['hour'] = df['date'].dt.hour
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['hour'], df['whitelight'], alpha=0.3)
plt.xlabel('Hour')
plt.ylabel('Whitelight')
plt.title('Whitelight vs. Hour')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(2000))
plt.xlim(-0.5, 24)
plt.show()
```
