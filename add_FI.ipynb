```python
#save the sleep period as list and mantaing the same format as the sleep scoring
sleep_period_list = []
for i in range(len(sleep_period)):
    sleep_period_list.append([sleep_period['start_datetime'][i], sleep_period['end_datetime'][i]])
```

```python
#save the sleep_scoring as list and mantaing the same format as the sleep scoring
sleep_scoring_list = []
for i in range(len(sleep_scoring)):
    sleep_scoring_list.append([sleep_scoring['datetime'][i], sleep_scoring['score'][i]])
```

```python
sleep_scoring_list
```

Fragmentation Index: The sum of the ‘Mobile time (%)’ and the ‘Immobile bouts <=1min (%)’. This is an indication of the degree of fragmentation of the sleep period, and can be used as an indication of sleep quality (or the lack of it).

```python
# Sample data for sleep_period and sleep_scori
sleep_period = pd.DataFrame(sleep_period_list, columns=['start_datetime', 'end_datetime'])
sleep_scoring = pd.DataFrame(sleep_scoring_list, columns=['datetime', 'score'])

# Convert datetime columns to datetime format
sleep_period['start_datetime'] = pd.to_datetime(sleep_period['start_datetime'])
sleep_period['end_datetime'] = pd.to_datetime(sleep_period['end_datetime'])
sleep_scoring['datetime'] = pd.to_datetime(sleep_scoring['datetime'])

# Initialize a list to store the results
results = []

# Calculate the amount of epochs scored as 0 for each sleep period
for i in range(len(sleep_period)):
    start = sleep_period['start_datetime'][i]
    end = sleep_period['end_datetime'][i]
    mask = (sleep_scoring['datetime'] >= start) & (sleep_scoring['datetime'] <= end)
    results.append(sleep_scoring.loc[mask, 'score'].value_counts().get(0, 0))

# Add the results to the sleep_period dataframe
sleep_period['score_0'] = results

```
